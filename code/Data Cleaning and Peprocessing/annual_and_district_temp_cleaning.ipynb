```python
"""Linear Regression
    Input Variable -> Data used for prediction
    Target Variable -> Data that is predicted
    
    y = mx + c
    where, y -> target variable
           x -> input variable
           m -> slope
           c -> intercept
           
    The second you plot the data on the graph calculation the mean the of 
    both input and target variable
    
    x-axis -> input variable
    y-axis -> target variable

    If the mean of x is p and the mean of y is q,
    plot x = p(vertical line) and y = q(horizontal line) on the graph.
    
    It is given that the line (of the graph) will pass through the intersection of
    the above plotted lines.
    
    For each data point in x, calculate x - mean(x)
    This gives us the deviation of x with its mean
    
    Do the same for y
    
    Calculate the square of x - mean(x)
    This ensures that positive and negative values do not nullify each other
    After calculating this for all values of x, calculate their sum
    
    Now calculate (x-mean(x))(y-mean(y))
    After calculating this for all values of x and y, calculate their sum
    
    ** m = E ((x-mean(X)(y-mean(y))) / E ((x-mean(x))^2)
    E = summation
    
    ** mean(y) = m * mean(x) + c
    Use this equation to calculate the value of c
    We already have m, mean(x) and mean(y)
    
    The difference between the points in the line that we plotted and the actual data points 
    is called Residual.
    Residual should be minimum
    
    Line of Best Fit: The kine we get is rotated left or right to find out the minimum error
    ... this line with minimum error is called the line of best fit
    
    We change the value of m and c to rotate the line
    Higher we rotate, more accurate line we get
    
    Error for one data point: Take the difference between the point on the line and
    the actual data point and square it
    To get the total error, take the sum of the squared value
    
    The rotation should not go beyond x = mean(x) and y = mean(y)
    
    R^2: Tells us how well the model predicts the actual data points
    
    Problem with Linear Regression: Outliers affect the result of the prediction negatively... Thus we have to
    remove them via EDA
    
    Ground Truth: Actual Data Points
    Actual Truth: Estimated Data Points
    
    Calculate the difference between actual data point and mean(y)
    Calculate the difference between estimated data point and mean(y)
    y hat (^ on top of y): value of y from y = mx+c for different values of x
    y hat -> Estimated Data Point
    ** R^2 = E (y hat - mean(y))^2 / E (y - mean(y))^2
    R^2 -> 1: Good Fit
    R^2 -> 0: Bad Fit
    
    Independent Columns -> Features (can be multiple)
    Target Column -> only one
    
    When the number of features increase, chances of overfitting also increses
    Overfitting -> when a model becomes too complex and "memorizes" the training data, failing to generalize and 
    make accurate predictions on new, unseen data. 
    Underfitting -> occurs when a model is too simple to capture the underlying patterns in the training data, 
    leading to poor performance on both the training and new, unseen data.
    
    Adjusted R^2 = 1 - (((1-R^2)(N-1)) / N-p-1)
    where, N -> Total Sample Size
           p -> Number of Features/Independent Variables"""
```

```python
#Applying Linear Regression for each district separately

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

feature_map = {
    "Jan-FebDistrict": ["JanFebNational", "AnnualNational"],
    "Mar-MayDistrict": ["MarMayNational", "AnnualNational"],
    "Jun-SepDistrict": ["JunSepNational", "AnnualNational"],
    "Oct-DecDistrict": ["OctDecNational", "AnnualNational"],
    "AnnualDistrict": ["JanFebNational", "MarMayNational", "JunSepNational", "OctDecNational", "AnnualNational"]
}

results = []

for district_name in monthly_district_temp["District"].unique():
    df_district = monthly_district_temp[monthly_district_temp["District"] == district_name]
    
    df_merged = df_district.merge(final_annual_temp, on="Year", how="inner")

    for target_col, features in feature_map.items():
        X = df_merged[features]
        y = df_merged[target_col]

        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)
        
        model = LinearRegression()
        model.fit(X_train, y_train)
        
        y_pred = model.predict(X_test)
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        
        results.append({
            "District": district_name,
            "Target": target_col,
            "Features": features,
            "RMSE": rmse
        })

results_df = pd.DataFrame(results)
```

```python
# Applying XGBoost for each district separately

from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

feature_map = {
    "Jan-FebDistrict": ["JanFebNational", "AnnualNational"],
    "Mar-MayDistrict": ["MarMayNational", "AnnualNational"],
    "Jun-SepDistrict": ["JunSepNational", "AnnualNational"],
    "Oct-DecDistrict": ["OctDecNational", "AnnualNational"],
    "AnnualDistrict": ["JanFebNational", "MarMayNational", "JunSepNational", "OctDecNational", "AnnualNational"]
}

results = []

for district_name in monthly_district_temp["District"].unique():
    df_district = monthly_district_temp[monthly_district_temp["District"] == district_name]
    
    df_merged = df_district.merge(final_annual_temp, on="Year", how="inner")

    for target_col, features in feature_map.items():
        X = df_merged[features]
        y = df_merged[target_col]

        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.2, shuffle=True, random_state=42
        )
        
        model = XGBRegressor(
            n_estimators=200,
            learning_rate=0.1,
            max_depth=4,
            subsample=0.8,
            colsample_bytree=0.8,
            random_state=42,
            n_jobs=-1
        )
        model.fit(X_train, y_train)
        
        y_pred = model.predict(X_test)
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        
        results.append({
            "District": district_name,
            "Target": target_col,
            "Features": features,
            "RMSE": rmse
        })

results_df = pd.DataFrame(results)
```

```python
import joblib
import os

MODEL_DIR = 'temp_imputation_models' 

def load_district_model(district, target_col):
    filename = f"{district}_{target_col}_xgb_model.joblib"
    filepath = os.path.join(MODEL_DIR, filename)
    
    if os.path.exists(filepath):
        return joblib.load(filepath)
    else:
        print(f"Error: Model file not found for {district} - {target_col}")
        return None
```

```python
rename_map = {
    'YEAR': 'Year',
    'ANNUAL': 'AnnualNational',
    'JAN-FEB': 'JanFebNational',
    'MAR-MAY': 'MarMayNational',
    'JUN-SEP': 'JunSepNational',
    'OCT-DEC': 'OctDecNational'
}
prediction_years = list(range(2014, 2020)) 
renamed_data = annual_temp.rename(columns=rename_map).copy()
renamed_data['Year'] = pd.to_numeric(renamed_data['Year'], errors='coerce').astype('int64')
prediction_temp = renamed_data[
    renamed_data['Year'].isin(prediction_years)
].reset_index(drop=True)

print("\nPrediction Temp (2014-2019) Head:")
print(prediction_temp.head())
```

```python
all_districts = monthly_district_temp['District'].unique()
national_cols_needed = ['Year', 'JanFebNational', 'MarMayNational', 
                        'JunSepNational', 'OctDecNational', 'AnnualNational']
prediction_df = pd.DataFrame({
    'Year': pd.Series(prediction_years).repeat(len(all_districts)),
    'District': pd.Series(all_districts).tolist() * len(prediction_years)
})
prediction_df = prediction_df.sort_values(by=['Year', 'District']).reset_index(drop=True)
prediction_df['Year'] = prediction_df['Year'].astype('int64')
national_data = prediction_temp[prediction_temp['Year'].isin(prediction_years)][national_cols_needed]
prediction_df_merged = pd.merge(
    prediction_df,
    national_data,
    on='Year',
    how='left' 
)
print("Prediction Dataframe Head (Features Ready for XGBoost):")
print(prediction_df_merged.head())
```

```python
import numpy as np
target_cols = [
    'Jan-FebDistrict', 
    'Mar-MayDistrict', 
    'Jun-SepDistrict', 
    'Oct-DecDistrict', 
    'AnnualDistrict'
]

final_prediction_input = prediction_df_merged.copy()
for col in target_cols:
    final_prediction_input[col] = np.nan

print("\nFirst five rows with new columns:")
print(final_prediction_input.head())
```

```python
import joblib
import os
import pandas as pd
from xgboost import XGBRegressor

MODEL_DIR = 'temp_imputation_models'
os.makedirs(MODEL_DIR, exist_ok=True)

feature_map = {
    "Jan-FebDistrict": ["Year", "JanFebNational", "AnnualNational"],
    "Mar-MayDistrict": ["Year", "MarMayNational", "AnnualNational"],
    "Jun-SepDistrict": ["Year", "JunSepNational", "AnnualNational"],
    "Oct-DecDistrict": ["Year", "OctDecNational", "AnnualNational"],
    "AnnualDistrict": ["Year", "JanFebNational", "MarMayNational", "JunSepNational", "OctDecNational", "AnnualNational"]
}

all_districts = monthly_district_temp['District'].unique() 
for district_name in all_districts:
    df_district = monthly_district_temp[monthly_district_temp["District"] == district_name].copy()
    df_merged = df_district.merge(final_annual_temp, on="Year", how="inner")

    for target_col, features in feature_map.items():
        if len(df_merged) < 2:
             continue
        
        X = df_merged[features]
        y = df_merged[target_col]

        model = XGBRegressor(
            n_estimators=200,
            learning_rate=0.1,
            max_depth=4,
            subsample=0.8,
            colsample_bytree=0.8,
            random_state=42,
            n_jobs=-1
        )
        model.fit(X, y)
        filename = f"{district_name}_{target_col}_xgb_model.joblib"
        filepath = os.path.join(MODEL_DIR, filename)
        
        joblib.dump(model, filepath) 
```

```python
national_temp_features = [
    'JanFebNational', 
    'MarMayNational', 
    'JunSepNational', 
    'OctDecNational', 
    'AnnualNational'
]

for col in national_temp_features:
    final_prediction_input[col] = pd.to_numeric(
        final_prediction_input[col], 
        errors='coerce'
    ).astype(np.float64)

for district in all_districts:
    district_data = final_prediction_input.loc[
        final_prediction_input['District'] == district
    ].copy()

    for target_col, features in feature_map.items():
        model = load_district_model(district, target_col)
        
        if model is not None:
            X_predict = district_data[features]
            predicted_values = model.predict(X_predict)
            final_prediction_input.loc[
                final_prediction_input['District'] == district, 
                target_col
            ] = predicted_values
```
