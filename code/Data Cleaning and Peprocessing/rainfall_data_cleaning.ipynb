```python
rainfall_df = rainfall_df[(rainfall_df['YEAR'] >= 1997) & (rainfall_df['YEAR'] <= 2017)]
columns_to_drop = ['JF', 'MAM', 'JJAS', 'OND']
rainfall_df = rainfall_df.drop(columns= columns_to_drop)
rainfall_df.head(1)
```

```python
rainfall_df = rainfall_df[rainfall_df['SUBDIVISION'] != 'Andaman & Nicobar Islands']
rainfall_df.reset_index()
rainfall_df.head()
```

```python
rainfall_df.isna().sum()
```

```python
rainfall_df.update(
    rainfall_df.groupby('SUBDIVISION').transform(lambda x: x.fillna(x.mean()))
)
```

```python
rainfall_df.isna().sum()
```

```python
season_month_mapping = {
    'Rabi': ['JAN', 'FEB', 'MAR'],                
    'Kharif': ['JUN', 'JUL', 'AUG', 'SEP'],      
    'Whole Year': ['ANNUAL'],                   
    'Autumn': ['OCT', 'NOV'],                     
    'Summer': ['APR', 'MAY'],                     
    'Winter': ['DEC', 'JAN', 'FEB']               
}

for season, months in season_month_mapping.items():
    valid_months = [col for col in months if col in rainfall_df.columns]
    rainfall_df[season] = rainfall_df[valid_months].sum(axis=1)

month_cols = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
rainfall_df.drop(columns=month_cols, inplace=True)
rainfall_df.rename(columns= {'SUBDIVISION' : 'State', 'YEAR' : 'Year'}, inplace= True)
print(rainfall_df.head())
```

```python
rainfall_df.sort_values(['State', 'Year'], inplace=True)

full_years = range(1997, 2020)

def fill_missing_years(group):
    group = group.set_index('Year').reindex(full_years)
    group['State'] = group['State'].ffill()  
    group = group.interpolate(method='linear')  
    return group.reset_index().rename(columns={'index': 'Year'})

rainfall_df = rainfall_df.groupby('State', group_keys=False).apply(fill_missing_years)
print(rainfall_df.groupby('State')['Year'].agg(['min', 'max']))

rainfall_df.reset_index(drop=True, inplace=True)
```

```python
rainfall_df['State'].unique()
```

```python
state_to_imd = {
    'Andhra Pradesh': ['Coastal Andhra Pradesh', 'Rayalseema'],
    'Arunachal Pradesh': ['Arunachal Pradesh'],
    'Assam': ['Assam & Meghalaya'],
    'Bihar': ['Bihar'],
    'Chhattisgarh': ['Chhattisgarh'],
    'Goa': ['Konkan & Goa'],
    'Gujarat': ['Gujarat Region', 'Saurashtra & Kutch'],
    'Haryana': ['Haryana Delhi & Chandigarh'],
    'Himachal Pradesh': ['Himachal Pradesh'],
    'Jammu and Kashmir': ['Jammu & Kashmir'],
    'Jharkhand': ['Jharkhand'],
    'Karnataka': ['Coastal Karnataka', 'North Interior Karnataka', 'South Interior Karnataka'],
    'Kerala': ['Kerala'],
    'Madhya Pradesh': ['East Madhya Pradesh', 'West Madhya Pradesh'],
    'Maharashtra': ['Madhya Maharashtra', 'Matathwada', 'Vidarbha'],
    'Manipur': ['Naga Mani Mizo Tripura'],  
    'Meghalaya': ['Assam & Meghalaya'],
    'Mizoram': ['Naga Mani Mizo Tripura'],
    'Nagaland': ['Naga Mani Mizo Tripura'],
    'Odisha': ['Orissa'],
    'Punjab': ['Punjab'],
    'Rajasthan': ['East Rajasthan', 'West Rajasthan'],
    'Sikkim': ['Sub Himalayan West Bengal & Sikkim'],
    'Tamil Nadu': ['Tamil Nadu'],
    'Telangana': ['Telangana'],
    'Tripura': ['Naga Mani Mizo Tripura'],
    'Uttar Pradesh': ['East Uttar Pradesh', 'West Uttar Pradesh', 'Haryana Delhi & Chandigarh'],  # includes Delhi
    'Uttarakhand': ['Uttarakhand'],
    'West Bengal': ['Gangetic West Bengal', 'Sub Himalayan West Bengal & Sikkim']
}

imd_to_state = {subdiv: state for state, subdivs in state_to_imd.items() for subdiv in subdivs}
rainfall_df['State'] = rainfall_df['State'].replace(imd_to_state)
rainfall_state_df = (
    rainfall_df
    .groupby(['State', 'Year'], as_index=False)
    .sum(numeric_only=True)
)
```

```python
len(rainfall_state_df['State'].unique())
```

```python
expected_states = [
    'Andhra Pradesh', 'Arunachal Pradesh', 'Assam', 'Bihar', 'Chhattisgarh', 'Goa', 
    'Gujarat', 'Haryana', 'Himachal Pradesh', 'Jammu and Kashmir', 'Jharkhand', 
    'Karnataka', 'Kerala', 'Madhya Pradesh', 'Maharashtra', 'Manipur', 'Meghalaya',
    'Mizoram', 'Nagaland', 'Odisha', 'Punjab', 'Rajasthan', 'Sikkim', 'Tamil Nadu',
    'Telangana', 'Tripura', 'Uttar Pradesh', 'Uttarakhand', 'West Bengal'
]

rainfall_states = rainfall_state_df['State'].unique().tolist()
missing_states = [s for s in expected_states if s not in rainfall_states]
print("Missing states in rainfall data:", missing_states)
```

```python
# Merge Telangana into Andhra Pradesh
rainfall_state_df['State'] = rainfall_state_df['State'].replace('Telangana', 'Andhra Pradesh')

rainfall_state_df = (
    rainfall_state_df
    .groupby(['State', 'Year'], as_index=False)
    .sum(numeric_only=True)
)
```

```python
proxy_mapping = {
    'Assam': 'Meghalaya',
    'Haryana': 'Punjab',  
    'Manipur': 'Tripura',
    'Mizoram': 'Tripura',
    'Nagaland': 'Tripura',
    'Sikkim': 'West Bengal'
}

for missing_state, proxy_state in proxy_mapping.items():
    if missing_state not in rainfall_state_df['State'].values and proxy_state in rainfall_state_df['State'].values:
        temp = rainfall_state_df[rainfall_state_df['State'] == proxy_state].copy()
        temp['State'] = missing_state
        rainfall_state_df = pd.concat([rainfall_state_df, temp], ignore_index=True)
```

```python
rainfall_state_df = rainfall_state_df[rainfall_state_df['State'] != 'Lakshadweep']
```
