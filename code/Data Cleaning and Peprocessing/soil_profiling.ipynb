```python
import pandas as pd
micronutrient = pd.read_csv(r'C:\DATA ANALYSIS AND DATA SCIENCE PROJECTS\AgriGrow\data\Raw Data\micronutrients_statewise.csv')
macronutrient = pd.read_csv(r'c:\DATA ANALYSIS AND DATA SCIENCE PROJECTS\AgriGrow\data\Raw Data\macronutrients_statewise.csv')
```

```python
soil_df = pd.merge(macronutrient, micronutrient, on='State', how='inner')
soil_df.head()
```

```python
soil_df.columns
```

```python
columns_to_drop =  ['Scheme_x', 'Cycle_x', 'Cu_Sufficient_x',
       'Cu_Deficient_x', 'Mn_Sufficient_x',
       'Mn_Deficient_x', 'Scheme_y', 'Cycle_y', 'n_High_y', 'n_Medium_y',
       'n_Low_y', 'p_High_y', 'p_Medium_y', 'p_Low_y', 'k_High_y',
       'k_Medium_y', 'k_Low_y', 'OC_High_y', 'OC_Medium_y', 'OC_Low_y',
       'pH_Alkaline_y', 'pH_Acidic_y', 'pH_Neutral_y', 'EC_NonSaline_y',
       'EC_Saline_y', 'S_Sufficient_y', 'S_Deficient_y', 'Fe_Sufficient_y',
       'Fe_Deficient_y', 'Zn_Sufficient_y', 'Zn_Deficient_y',
       'Cu_Sufficient_y', 'Cu_Deficient_y', 'B_Sufficient_y', 'B_Deficient_y',
       'Mn_Sufficient_y', 'Mn_Deficient_y']

soil_df.drop(columns= columns_to_drop, inplace= True)
```

```python
soil_df.columns
```

```python
rename_dict = {
    'State': 'State',
    
    'n_High_x': 'Nitrogen_High_%',
    'n_Medium_x': 'Nitrogen_Medium_%',
    'n_Low_x': 'Nitrogen_Low_%',
    'p_High_x': 'Phosphorus_High_%',
    'p_Medium_x': 'Phosphorus_Medium_%',
    'p_Low_x': 'Phosphorus_Low_%',
    'k_High_x': 'Potassium_High_%',
    'k_Medium_x': 'Potassium_Medium_%',
    'k_Low_x': 'Potassium_Low_%',
    
    'OC_High_x': 'Organic_Carbon_High_%',
    'OC_Medium_x': 'Organic_Carbon_Medium_%',
    'OC_Low_x': 'Organic_Carbon_Low_%',
    'pH_Alkaline_x': 'pH_Alkaline_%',
    'pH_Acidic_x': 'pH_Acidic_%',
    'pH_Neutral_x': 'pH_Neutral_%',
    'EC_NonSaline_x': 'EC_NonSaline_%',
    'EC_Saline_x': 'EC_Saline_%',
    
    'S_Sufficient_x': 'Sulphur_Sufficient_%',
    'S_Deficient_x': 'Sulphur_Deficient_%',
    'Fe_Sufficient_x': 'Iron_Sufficient_%',
    'Fe_Deficient_x': 'Iron_Deficient_%',
    'Zn_Sufficient_x': 'Zinc_Sufficient_%',
    'Zn_Deficient_x': 'Zinc_Deficient_%',
    'B_Sufficient_x': 'Boron_Sufficient_%',
    'B_Deficient_x': 'Boron_Deficient_%'
}

soil_df.rename(columns=rename_dict, inplace=True)
```

```python
soil_df['State'] = soil_df['State'].str.title().str.strip()
soil_df = soil_df[~soil_df['State'].isin(['Andaman & Nicobar Islands', 'Ladakh'])]
print(type(soil_df['Sulphur_Sufficient_%'][0]))
```
